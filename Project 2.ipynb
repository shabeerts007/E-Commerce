```python
data['InvoiceDate'] = pd.to_datetime(data.InvoiceDate, format='%m/%d/%Y %H:%M')
```

```python
data.info()
```

```python
data.head()
```

```python
# Add the columns - Month, Day and Hour for the invoice

data.insert(loc=2, column='year_month', value=data['InvoiceDate'].map(lambda x: 100*x.year + x.month))
data.insert(loc=3, column='month', value=data.InvoiceDate.dt.month)
# +1 to make Monday=1.....until Sunday=7
data.insert(loc=4, column='day', value=(data.InvoiceDate.dt.dayofweek)+1)
data.insert(loc=5, column='hour', value=data.InvoiceDate.dt.hour)
```

```python
data2 = data.dropna()
```

```python
data2.info()
```

```python
duplicate4=data2.duplicated()
```

```python
duplicate4.value_counts()
```

```python
# Remove duplicate rows

data2.drop_duplicates(subset=None, keep='first', inplace=False)
```

```python
# Remove rows which have negative values in Quantity column

data3= data2[data2.Quantity > 0]
```

```python
data3.describe()
```

```python
# How many order made by the customers?

data3.groupby(by=['CustomerID','Country'], as_index=False)['InvoiceNo'].count().head(10)
```

```python
data3
```

```python
data3['InvoiceNo'].unique()
```

```python
data3['CustomerID'].unique()
```

```python
orders = data3.groupby(by=['CustomerID','Country'], as_index=False)['InvoiceNo'].count()

plt.subplots(figsize=(15,6))
plt.plot(orders.CustomerID, orders.InvoiceNo)
plt.xlabel('Customers ID')
plt.ylabel('Number of Orders')
plt.title('Number of Orders for different Customers')
plt.show()
```

```python
# TOP 5 customers with higher number of orders


print('The top 5 customers with most number of orders')
orders.sort_values(by='InvoiceNo', ascending=False).head()
```

```python
data3['Amount_spent'] = data3['Quantity'] * data3['UnitPrice']
```

```python
# How much money spent by the customers?

data3
```

```python
money_spent = data3.groupby(by=['CustomerID','Country'], as_index=False)['Amount_spent'].sum()

plt.subplots(figsize=(15,6))
plt.plot(money_spent.CustomerID, money_spent.Amount_spent)
plt.xlabel('Customers ID')
plt.ylabel('Money spent ($)')
plt.title('Money Spent by different customers')
plt.show()
```

```python
# TOP 5 customers with highest money spent

print('The top 5 customers with highest money spent')
money_spent.sort_values(by='Amount_spent', ascending=False).head()
```

```python
# How many orders per month

data3.groupby('InvoiceNo')['month'].unique().value_counts().sort_index()
```

```python
month_order = data3.groupby('InvoiceNo')['year_month'].unique().value_counts().sort_index().plot(kind='bar',figsize=(15,10),color=color[0])
month_order.set_xlabel('Month',fontsize=15)
month_order.set_ylabel('Number of Orders',fontsize=15)
month_order.set_title('Number of orders for different Months (1st Dec 2010 - 9th Dec 2011)',fontsize=15)
month_order.set_xticklabels(('Dec_10','Jan_11','Feb_11','Mar_11','Apr_11','May_11','Jun_11','July_11','Aug_11','Sep_11','Oct_11','Nov_11','Dec_11'), rotation='horizontal', fontsize=13)
plt.show()
```

```python
# How many orders per day

data3.groupby('InvoiceNo')['day'].unique().value_counts().sort_index()
```

```python
day_order = data3.groupby('InvoiceNo')['day'].unique().value_counts().sort_index().plot(kind='bar',figsize=(15,10),color=color[0])
day_order.set_xlabel('Month',fontsize=15)
day_order.set_ylabel('Number of Orders',fontsize=15)
day_order.set_title('Number of orders for different days',fontsize=15)
day_order.set_xticklabels(('Mon','Tue','Wed','Thur','Fri','Sun'), rotation='horizontal', fontsize=13)
plt.show()
```

```python
# How many orders for each country?

country_orders = data3.groupby('Country')['InvoiceNo'].count().sort_values()
country_orders
```

```python

plt.subplots(figsize=(15,8))
country_orders.plot(kind='barh', fontsize=12,color=color[0])
plt.xlabel('Number of Orders', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.title('Number of Orders for different Countries', fontsize=12)
plt.show()
```

```python
# How much money spent by each country?


country_amount_spent = data3.groupby('Country')['Amount_spent'].sum().sort_values()
country_amount_spent
```

```python
plt.subplots(figsize=(15,10))
country_amount_spent.plot(kind='barh', fontsize=12,color=color[0])
plt.xlabel('Money spent (Dollar in millions)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.title('Money Spent by different countries', fontsize=12)
plt.show()
```
